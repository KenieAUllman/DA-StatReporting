Just from looking at my data, I can see that there is a 2.1% difference in fruit/veggie consumption between not overweight teens and overweight teens. That doesn't seem highly significicant. However, there is a 7.5% difference in consumption between teens who don't regularly work out vs teens who do. 

```python
teen_income = select_AgeGroup_Adolescent.loc[(select_AgeGroup_Adolescent["Category"] == "Income")]
teen_income
```

```python
fig2 = px.bar(teen_income, x="Type", color="Servings of F/V => 5")
fig2.show()
```

The teens who eat the highest percentage of fruits/veggies have a family income of $25,000 - $34,999. 
The teens who eat the smallest percentage of fruits/veggies have a family income of $50,000 - $74,000. 

```python
teen_gen_by_age = select_AgeGroup_Adolescent.iloc[3:9]
teen_gen_by_age
```

```python
fig3 = px.scatter(teen_gen_by_age, x="Type", y="Servings of F/V => 5", color="Category")
fig3.show()
```

The teens in this study who eat the most veggies are females between the age of 12-13. The ones who eat the least are boys between the ages of 16-17.
